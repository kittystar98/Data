```python
#      CSV-    
import pandas as pd
import matplotlib.pyplot as plt

#    CSV-   DataFrame
df = pd.read_csv('breed_traits.csv')

#       
breed1 = 'Labrador'
breed2 = 'Rottweilers'
features = ['Affectionate_With_Family', 'Good_With_Young_Children', 'Good_With_Other_Dogs']
features2 = ['Openness_To_Strangers', 'Adaptability_Level']
breeds = ['Labrador', 'Rottweilers']
columns = ['Affectionate_With_Family', 'Good_With_Young_Children', 'Good_With_Other_Dogs','Openness_To_Strangers', 'Adaptability_Level']


#       
subset1 = df[df['Breed'].isin(breeds)][features]
averages1 = subset1.sum(axis=1)/3
subset2 = df[df['Breed'].isin(breeds)][features2]
averages2 = subset2.sum(axis=1)/2
subset3 = df[df['Breed'].isin(breeds)][columns]
averages3 = subset3.sum(axis=1)/5

#       
values_breed1 = []
values_breed2 = []
values_breed3 = []
values_breed4 = []

#   axis
fig, (ax1, ax2, ax3,ax4,ax5) = plt.subplots(nrows=1, ncols=5, figsize=(50, 5))

#    
for feature in features:
    values_breed1.append(df[(df['Breed'] == breed1)][feature].values[0])
    values_breed2.append(df[(df['Breed'] == breed2)][feature].values[0])

ax1.bar([i-0.2 for i in range(len(features))], values_breed1, width=0.4, label=breed1)
ax1.bar([i+0.2 for i in range(len(features))], values_breed2, width=0.4, label=breed2)

#     
ax1.set_title(f'  : {breed1} vs. {breed2}')
ax1.set_xlabel('')
ax1.set_ylabel('')

#     x      
ax1.set_xticks([i for i in range(len(features))])
ax1.set_xticklabels(features)

#  
ax1.legend()
#   
for feature in features2:
    values_breed3.append(df[(df['Breed'] == breed1)][feature].values[0])
    values_breed4.append(df[(df['Breed'] == breed2)][feature].values[0])

ax3.bar([i-0.2 for i in range(len(features2))], values_breed3, width=0.4, label=breed1)
ax3.bar([i+0.2 for i in range(len(features2))], values_breed4, width=0.4, label=breed2)

#     
ax3.set_title(f'  : {breed1} vs. {breed2}')
ax3.set_xlabel('')
ax3.set_ylabel('')

#     x      
ax3.set_xticks([i for i in range(len(features2))])
ax3.set_xticklabels(features2)

#  
ax3.legend()

#      
ax2.bar(breeds, averages1)

#     
ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_title('  1 ')

#      
ax4.bar(breeds, averages2)

#     
ax4.set_xlabel('')
ax4.set_ylabel('')
ax4.set_title('  3 ')

#     
ax5.bar(breeds, averages3)

#     
ax5.set_xlabel('')
ax5.set_ylabel('')
ax5.set_title('     ')

#    
plt.show()
```
